```python
# Problem 1 Using Python:
# The couple of lines below defines n as 3, along with generating a into squares and b into cubes
# to create integers and print them into a list.
n = 3
a = [i**2 for i in range (n+1)]
b = [i**3 for i in range (n+1)]
# The outcomes of the definements of both the creations of Vectors a & b.
print("Vector a:", a)
print("Vector b:", b)
```

```python
# Code Attribution for Problem 2:
# ; Title: Assignment 2.2
# ; Author: Armando Fandango
# ; Date: 08 December 2024
# ; Modified By: <Faith Wade>
# ; Description: The programs purpose is to print the iris dataset in a description &
# ;              plot column 1 and colum 2 to create a visualization of the dataset.

# Title: Python-Data-Analysis / Chapter 01: Getting Started
# Author: Armando Fandango
# Date: 08 December 2024
# Modified By: <Faith Wade>
# Description: The program was modified to create a print of the iris dataset, along with creating
#              a scatter plot to define the sepal width and sepal length of columns 1 & 2.

# Source: https://github.com/fwade0572/Python-Data-Analysis/blob/master/Chapter01/ch-01.ipynb
```

```python
# Problem 2:
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
# The iris dataset is loaded into the programs and from there commands panadas to read the dataframe of its attributes.  From there
# the taget column is analyzed and added to the dataframe where it needs to go.
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns = iris.feature_names)
iris_df['target'] = iris.target
# The iris dataset is printed into a description to list the data that it contains.
Description = iris.DESCR
print(Description)
# The data from the iris dataset is plotted into a scatterplot graph to join together columns 1 & 2
# to show the speal width and sepal length together by creating different additions to create a deeper visual.
plt.figure(figsize=(8, 6))
plt.scatter(iris_df[iris.feature_names[0]], iris_df[iris.feature_names[1]], c=iris_df['target'], cmap='viridis')
plt.colorbar(label='Target')
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.grid(True)
plt.show()
```

```python
# Problem 3: Could not work with Boston Dataset so I used the Calfornia Housing one instead.
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
# The program loads the dataset into the system and then creates a description function to
# display the dataset to present a visual.
california = fetch_california_housing(as_frame=True)
california_df = california.frame
Description = california.DESCR
print(Description)
# A visual plot is shown to emphasize the dataset between the AveOccup and MedInc variables using certain
# additions and specifications.  (I had a little trouble with the scaling the points on this one, so if you have any tips
# on how to improve or any sample lines of code you would recommend I'm all ears).
plt.figure(figsize=(8, 6))
plt.scatter(california_df['AveOccup'] + np.random.normal(0, 50, len(california_df)), california_df['MedInc'] +
            np.random.normal(0, 100, len(california_df)), marker='+', color='blue')
plt.title('Average Occupants per Household vs. Median Income')
plt.xlabel('AveOccup')
plt.ylabel('MedInc')
plt.grid(True)
plt.show()
```

```python
# Code Attribution for Problem 7:
# ; Title: Assignment 2.2
# ; Author: Armando Fandango
# ; Date: 08 December 2024
# ; Modified By: <Faith Wade>
# ; Description: The purpose of this program is to define an array, select a part of the indexes,
# ;              extract specific elements and reversing the array.

# Title: Python-Data-Analysis / Chapter 02: NumPy
# Author: Armando Fandango
# Date: 08 December 2024
# Modified By: <Faith Wade>
# Description: The program was modified to define an array containing numbers all the way up to 8, select a part
#              of the indexes from 3-7, extracing elements 3-6, and reversing the array up through 8.

# Source: https://github.com/fwade0572/Python-Data-Analysis/blob/master/Chapter02/ch-02.ipynb
```

```python
# Problem 7:
import numpy as np
# Setting the array up to the number wanted (in this case including 8).
array = np.arange(9)
# Indexes 3-7 selected to be shown.
part_array = array[3:7]
# Elements from 0-7 shown in increments of 2.
incremented_array = array[0:8:2]
# array[::-1] used to reverse the array provided.
reversed_array = array[::-1]

print(array)
print(part_array)
print(incremented_array)
print(reversed_array)
```

```python
# Code Attribution for Problem 8:
# ; Title: Assignment 2.2
# ; Author: Armando Fandango
# ; Date: 08 December 2024
# ; Modified By: <Faith Wade>
# ; Description: The programs purpose is to show the functions and how they can work with an array
# ;              provided below.

# Title: Python-Data-Analysis / Chapter 02: NumPy
# Author: Armando Fandango
# Date: 08 December 2024
# Modified By: <Faith Wade>
# Description: The program was modified to demonstrate how the functions of ravel, flatten,
#              transpose, resize, and setting the shape with a tuple to show the different visuals it can provide.

# Source: https://github.com/fwade0572/Python-Data-Analysis/blob/master/Chapter02/ch-02.ipynb
```

```python
# Problem 8:
import numpy as np
# setting up the rray wanting to be used.
b = np.arange(30).reshape(5, 6)
print(b)
# Using the ravel function through the operation in the textbook.
print(b.ravel())
# Using the flatten function through the operation in the textbook.
print(b.flatten())
#  Setting the shape with a tuple through the operation in the textbook.
b.shape = (6, 5)
print(b)
# Using the transpose function through the operation in the textbook.
print(b.transpose())
# Using the resize function through the operation in the textbook.
b.resize((3, 10))
print(b)
```

```python
# Code Attribution for Problem 9:
# ; Title: Assignment 2.2
# ; Author: Armando Fandango
# ; Date: 08 December 2024
# ; Modified By: <Faith Wade>
# ; Description: The programs purpose is to use certain functions to create an array and calculate its
# ;              Number of Dimensions, Count of Elements, Count of Bytes, and Full Count of Bytes.

# Title: Python-Data-Analysis / Chapter 02: NumPy
# Author: Armando Fandango
# Date: 08 December 2024
# Modified By: <Faith Wade>
# Description: The program was modified to demonstrate the calculations made to create an ocerall analyzation
#              of the array provided given the b.ndim, b.size, b.itemsize, & b.nbytes functions.

# Source: https://github.com/fwade0572/Python-Data-Analysis/blob/master/Chapter02/ch-02.ipynb
```

```python
# Problem 9:
import numpy as np
# Setting up the array wanting to be used.
b = np.arange(30).reshape(5, 6)
print(b)
# Using the b.ndim, b.size, b.itemsize, b.nbytes, and multiplying b.size and b.itemsize
# to get an overall anayzation of the array.
print("Number of Dimensions:", b.ndim)
print("Count of Elements", b.size)
print("Count of Bytes", b.itemsize)
print("Full Count of Bytes:", b.nbytes)
print(b.size * b.itemsize)
```

```python
# Code Attribution for Problem 10:
# ; Title: Assignment 2.2
# ; Author: Armando Fandango
# ; Date: 08 December 2024
# ; Modified By: <Faith Wade>
# ; Description: The programs purpose is to use an image to be able to take an array and
# ;              convert it into a python list.

# Title: Python-Data-Analysis / Chapter 02: NumPy
# Author: Armando Fandango
# Date: 08 December 2024
# Modified By: <Faith Wade>
# Description: The progoram was modified to be able to work with the Google.Colab format and demonstrate
#              the creation of turning the array into a python list and work with an original, copied, and modified image.

# Source: https://github.com/fwade0572/Python-Data-Analysis/blob/master/Chapter02/ch-02.ipynb
```

```python
# Problem 10:
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc
# Setting up the array wanting to be worked with along with classifying its list.
array = np.array([1, 2, 3, 4, 5])
python_list = array.tolist()
# Printing the oucomes and sequences form the lines of code above.
print(array)
print(python_list)
print(type(python_list))
# Uzing the scipy library to create an image ot use for further lines of code below,
# setting all the commands.
face = misc.face()
acopy = face.copy()
aview = face.copy()
aview.flat = 0
# Creating the original image provided through the textbook code.
plt.figure(figsize=(10, 8))
plt.subplot(221)
plt.title("Original Image")
plt.imshow(face)
# Creating a Copy of the Image provided through the textbook code.
plt.subplot(222)
plt.title("Copy of Image")
plt.imshow(acopy)
# Creating a modified image provided through the textbook code.
plt.subplot(223)
plt.title("Modified Image")
plt.imshow(aview)
# Presenting the results in a visualization all together.
plt.tight_layout()
plt.show()
```
