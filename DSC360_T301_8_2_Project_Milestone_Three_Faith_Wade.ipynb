# Bag of Words Modeling
This program is made for the inent to study the mental health corpus provided and to build a general vocabulary while creating frequences to detect any specific values in a binary order.

```python
# Installing/importing the needed packages & modules for creating a Text Normalizer that
# works for what is needed in the assignment for code blocks below.
!pip install --upgrade nltk
import nltk
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize.treebank import TreebankWordTokenizer
from sklearn.feature_extraction.text import CountVectorizer
nltk.download('stopwords')
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("mental_health.csv")
# Identifying the classes of different parts of the normalizer to specify which kind of
# cleaning functions need to occur by pasing a pandas series and using lambda for the
# cleaning functions.
class Text_Normalizer:
  def __init__(self, remove_stopwords=False):
      self.remove_stopwords = remove_stopwords
      self.tokenizer = TreebankWordTokenizer()
      if remove_stopwords:
          self.stop_words = set(stopwords.words('english'))
  def remove_html_tags(self, text):
      return re.sub(r'<[^>]+>', '', text)
  def to_lowercase(self, text):
      return text.lower()
  def tokenize(self, text):
      return self.tokenizer.tokenize(text)
  def remove_special_tokens(self, tokens):
      return[
          token for token in tokens
          if token.isalpha() and (not self.remove_stopwords or token not in self.stop_words)
      ]
  def detokenize(self, tokens):
      return ' '.join(tokens)
  def remove_whitespace(self, text):
      return re.sub(r'\s+', ' ', text).strip()
  def normalize_corpus(self, text_series: pd.Series):
      return(
          text_series
          .apply(self.remove_html_tags)
          .apply(self.to_lowercase)
          .apply(self.tokenize)
          .apply(self.remove_special_tokens)
          .apply(lambda tokens: [t for t in tokens if len(t) > 2])
          .apply(self.detokenize)
          .apply(self.remove_whitespace)
      )
# Taking the text normalizer from above from week 4 and incoproating it into the data necessary.
text_col = 'text'
normalizer = Text_Normalizer(remove_stopwords=True)
norm_corpus = normalizer.normalize_corpus(df[text_col])
# Creating the base format for the Bag of Words Model:
cv = CountVectorizer(min_df=0., max_df=1.)
cv_matrix = cv.fit_transform(norm_corpus)
cv_matrix = cv_matrix.toarray()
vocab = cv.get_feature_names_out()
# Creating the DataFrame and physical result for the model.
bow_df = pd.DataFrame(cv_matrix, columns=vocab)
bow_df['label'] = df['label']
bow_df.head()
```

# Bag of N-Grams Modeling
The Bag of N-Grams Model program is made to analyze the Mental Health Corpus and analyze the sequence of the words in general to draw inferences at how deep sentimental analysis plays in this textual data.

```python
# Setting up the format for the Bag of N-Grams Model using its range:
bv = CountVectorizer(ngram_range=(2,2))
bv_matrix = bv.fit_transform(norm_corpus)
bv_matrix = bv_matrix.toarray()
vocab = bv.get_feature_names_out()
# Creating a physical DataFrame for a condensed array of results.
ngrams_df = pd.DataFrame(bv_matrix, columns=vocab)
ngrams_df['label'] = df['label']
ngrams_df.head()
```

# TF-IDF Modeling
This program is intended to measure out and detect frequencies at how often certain words appear inside of the Mental Health Corpus to create an overall data flow to determine if the source includes more positive, negative, and/or neutral social media posts.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
# Applying the normalizer to the code in order to format the TF-IDF Model.
normalizer = Text_Normalizer(remove_stopwords=True)
norm_corpus = normalizer.normalize_corpus(df['text'])
# Since the corpus was so big, I had to narrow it down to the first 1000 features in order
# for an output to come through.
tv = TfidfVectorizer(
    min_df=0., max_df=1., norm='l2',
    use_idf=True, smooth_idf=True,
    max_features=1000
)
tv_matrix = tv.fit_transform(norm_corpus).toarray()
vocab = tv.get_feature_names_out()
# Printing results.
tfidf_df = pd.DataFrame(np.round(tv_matrix, 2), columns=vocab)
tfidf_df.head()
```
